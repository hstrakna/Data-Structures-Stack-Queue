```python
def maxSubCubic(array, n):
    maxSum = 0  

    for i in range(0,n-1):
        for j in range(i,n-1):
            sum = 0
            for k in range(i,j):
                sum = sum + array[k]
                if sum > maxSum:
                    maxSum = sum                
    return maxSum
```

```python
def maxSubQuadratic(array):
    n = len(array)
    if n == 0:
        return
    else:
        sums = []
        temp = 0
        for i in range(0,n):
            temp += array[i]
            sums.append(temp)
        for i in range(0, n):
            for j in range(0, i):
                subsequence = sums[i] - sums[j]
                sums.append(subsequence)
        
    return max(sums)
```

```python
def findMaxSum(array, left, mid, right):
 
    tempSum = 0
    leftSum = 0
    # Loop backwards to make sure that we get the right answer for a sum that straddles left and right subarrays
    for i in range(mid, left-1, -1):
        tempSum = tempSum + array[i]
        if (tempSum > leftSum):
            leftSum = tempSum
 
    tempSum = 0
    rightSum = 0
    for i in range(mid+1, right):
        tempSum = tempSum + array[i]
        if (tempSum > rightSum):
            rightSum = tempSum
 
    return max(leftSum + rightSum, leftSum, rightSum)
```

```python

def maxSubNLogN(array, left, right):    
    
    mid = (left+right)//2
    
    if (left == right):
        return array[left]
 
    return max(maxSubNLogN(array, left, mid),
               maxSubNLogN(array, mid+1, right),
               findMaxSum(array, left, mid, right)) 
```

```python
#Create Stack class
class Stack:
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def peek(self):
        return self.items[len(self.items)-1]
    def size(self):
        return len(self.items)
```

```python
#Read maze from text file
def readMaze(m, filename):
    mazeFile = open(filename, "r")
    lines = mazeFile.readlines()
    for line in lines:
        line = line.strip()
        row = [c for c in line]
        m.append(row)
```

```python
#Function to determine if move is valid (0 or E) or False (1)
def isValidMove(position):
    (row,col) = position
    if col < 0 or row < 0:
        return False
    if col >= len(maze[0]) or row >= len(maze):
        return False
    if maze[row][col] == '0' or maze[row][col] == 'E':
        return True
    else:
        return False
```

```python
#Function to solve the maze
def solveMaze(maze, start):
    s = Stack()
    (startRow, startCol) = start
    if maze[startRow][startCol] == '1':
        print("Please select a valid starting point.")
    else:
        s.push((startRow, startCol))
    while s.isEmpty() != True:
        print("Now visiting: {}".format(s.peek()))
        (row, col) = s.pop()
        
        #Break if we reach E
        if maze[row][col] == 'E':
            print("Maze solved!")
            break
        
        #Mark start
        if (row, col) == (startRow, startCol):
            maze[row][col] = 'S'
        #Mark each visited position
        else:
            maze[row][col] = '+'
            
        #Check if moves are valid and push valid moves onto stack
        if isValidMove((row+1, col)) == True:
            s.push((row+1,col))
        if isValidMove((row, col+1)) == True:
            s.push((row,col+1))
        if isValidMove((row-1, col)) == True:
            s.push((row-1,col))
        if isValidMove((row, col-1)) == True:
            s.push((row,col-1))

        #Alert user that we are stuck when there are no valid moves left
        if s.isEmpty() == True:
            maze[row][col] = '!'
            print("I'm stuck!")
```
